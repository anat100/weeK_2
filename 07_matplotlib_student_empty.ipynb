## Add elements to the figure
### After creating the plot you can customize it using additional functions.

- Let's add a title and label the axes.
- Put all functions for a plot into the same cell of a Jupyter Notebook before plt.show():

```python
x_vals = range(1,8) 
y_vals = [12,34,23,98,40,45,18]

# plotting starts

plt.style.use('ggplot')
plt.figure(figsize=(6.4, 4.8))

plt.plot(x_vals, y_vals) # here you'll put in your plutting values (x,y)

# title
plt.title('Juniper Graph')

# axes labels
plt.xlabel('A')
plt.ylabel('B')


plt.show() 
```

```python
## changing rhe axis range (or starting my plot from '0')

_vals = range(1,8) 
y_vals = [12,34,23,98,40,45,18]

# plotting starts

plt.style.use('ggplot')
plt.figure(figsize=(6.4, 4.8))

plt.plot(x_vals, y_vals) # here you'll put in your plutting values (x,y)

# title
plt.title('Juniper Graph')

# axes labels
plt.xlabel('A')
plt.ylabel('B')


plt.xlim([0,110]) # by chaning plt.ylim or plt.xlin - I can change the display of my graph
plt.show()
```

## Customizing the plotted data
- The functions plt.scatter, plt.hist, plt.bar and plt.plot accept many input parameter that can be used to fine-tune the plotted data.

## Markers
- Determine the shape of the plotted points. Try out: v, ., o or s.

```python
# create a scatter plot

plt.figure(figsize=(9, 3))

plt.scatter(x_vals, y_vals, marker='X'); # if you want to have any other plot than the default, you need to redefine it with this function 

# 's' stands for the marker's shape that it will use in my plot
```

## Colors

- You can provide a single color

```python
# create a bar plot and change it's color! 

plt.figure(figsize=(9, 3))

plt.scatter(x_vals, y_vals, color='teal');

# stacked=true  if you want to add more info and stack it
```

```python
# same but with a list of colors

plt.figure(figsize=(9, 3))

# using 'bar'
plt.bar(x_vals, y_vals, color=['purple','teal']);
```

```python
# changing the size or width of my plot

plt.figure(figsize=(9, 3))

plt.bar(x_vals, y_vals, color=['purple','teal'], width=0.2);
```

```python
plt.figure(figsize=(9, 3))

# changing marker size

plt.scatter(x_vals, y_vals, marker='P', s=500);
```
